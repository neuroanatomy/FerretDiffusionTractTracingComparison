## Correlation with weight

```python
def removeWeights(ref,d1,d2,d3,d4,d5,d6,maxmode):  ##maxmode = 1 (remove strongest weigth), 0 (weakest)
    correl = np.zeros([round(len(ref)/2)+1,12])
    
    ref_tmp = np.copy(ref)
    
    d1_tmp = np.copy(d1)
    d2_tmp = np.copy(d2)
    d3_tmp = np.copy(d3)
    d4_tmp = np.copy(d4)
    d5_tmp = np.copy(d5)
    d6_tmp = np.copy(d6)
    
    i = 0
    while i <= round(len(ref)/2):
        correl[i,0] = sc.spearmanr(ref_tmp,d1_tmp).correlation
        correl[i,1] = sc.spearmanr(ref_tmp,d1_tmp).pvalue
        correl[i,2] = sc.spearmanr(ref_tmp,d2_tmp).correlation
        correl[i,3] = sc.spearmanr(ref_tmp,d2_tmp).pvalue
        correl[i,4] = sc.spearmanr(ref_tmp,d3_tmp).correlation
        correl[i,5] = sc.spearmanr(ref_tmp,d3_tmp).pvalue
        correl[i,6] = sc.spearmanr(ref_tmp,d4_tmp).correlation
        correl[i,7] = sc.spearmanr(ref_tmp,d4_tmp).pvalue
        correl[i,8] = sc.spearmanr(ref_tmp,d5_tmp).correlation
        correl[i,9] = sc.spearmanr(ref_tmp,d5_tmp).pvalue
        correl[i,10] = sc.spearmanr(ref_tmp,d6_tmp).correlation
        correl[i,11] = sc.spearmanr(ref_tmp,d6_tmp).pvalue
        i += 1
        
        if maxmode:
            mode = np.argmax(ref_tmp)
        else:
            mode = np.argmin(ref_tmp)
            
        ref_tmp = np.delete(ref_tmp,mode)
        d1_tmp = np.delete(d1_tmp,mode)
        d2_tmp = np.delete(d2_tmp,mode)
        d3_tmp = np.delete(d3_tmp,mode)
        d4_tmp = np.delete(d4_tmp,mode)
        d5_tmp = np.delete(d5_tmp,mode)
        d6_tmp = np.delete(d6_tmp,mode)
        
    return correl
```

```python
def removeWeights_pearson(ref,d1,d2,d3,d4,d5,d6,maxmode):  ##maxmode = 1 (remove strongest weigth), 0 (weakest)
    correl = np.zeros([round(len(ref)/2)+1,12])
    
    ref_tmp = np.copy(ref)
    
    d1_tmp = np.copy(d1)
    d2_tmp = np.copy(d2)
    d3_tmp = np.copy(d3)
    d4_tmp = np.copy(d4)
    d5_tmp = np.copy(d5)
    d6_tmp = np.copy(d6)
    
    i = 0
    while i <= round(len(ref)/2):
        correl[i,0] = sc.pearsonr(ref_tmp,d1_tmp)[0]
        correl[i,1] = sc.pearsonr(ref_tmp,d1_tmp)[1]
        correl[i,2] = sc.pearsonr(ref_tmp,d2_tmp)[0]
        correl[i,3] = sc.pearsonr(ref_tmp,d2_tmp)[1]
        correl[i,4] = sc.pearsonr(ref_tmp,d3_tmp)[0]
        correl[i,5] = sc.pearsonr(ref_tmp,d3_tmp)[1]
        correl[i,6] = sc.pearsonr(ref_tmp,d4_tmp)[0]
        correl[i,7] = sc.pearsonr(ref_tmp,d4_tmp)[1]
        correl[i,8] = sc.pearsonr(ref_tmp,d5_tmp)[0]
        correl[i,9] = sc.pearsonr(ref_tmp,d5_tmp)[1]
        correl[i,10] = sc.pearsonr(ref_tmp,d6_tmp)[0]
        correl[i,11] = sc.pearsonr(ref_tmp,d6_tmp)[1]
        i += 1
        
        if maxmode:
            mode = np.argmax(ref_tmp)
        else:
            mode = np.argmin(ref_tmp)
            
        ref_tmp = np.delete(ref_tmp,mode)
        d1_tmp = np.delete(d1_tmp,mode)
        d2_tmp = np.delete(d2_tmp,mode)
        d3_tmp = np.delete(d3_tmp,mode)
        d4_tmp = np.delete(d4_tmp,mode)
        d5_tmp = np.delete(d5_tmp,mode)
        d6_tmp = np.delete(d6_tmp,mode)
        
    return correl
```
